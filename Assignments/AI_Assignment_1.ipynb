.2. Write a function to Calculate Area and Perimeter of Rectangle

```python
def rectangle(l,b):
  area = l * b
  perimeter = 2 * (l+ b)
  return area, perimeter
l = float(input("Enter the length of the rectangle: "))
b = float(input("Enter the breadth of the rectangle: "))
area, perimeter = rectangle(l,b)
print("The area of the rectangle is:", area, "sq.units")
print("The perimeter of the rectangle is:", perimeter, "sq.units")
```


3. Write a python Program to find out given No is Prime or Not.

```python
try:
  x=int(input("Enter any positive integer: "))
  if x>0:
    try:
      if x==1:
        print("It is neither prime nor composite.")
      if x==2:
        print("It is a prime number.")
      for i in range(2,x):
        if (x % i) == 0:
          print("It is composite number.")
          break
      else:
        print("It is prime number.")
    except ValueError:
      print("Invalid Input.")
  else:
    print("Enter positive integer.")
except ValueError:
      print("Invalid Input.")
```

4. Write a Python program that prompts the user to input three numbers and then determines and
prints the largest among them.

First 100 units  no charge
Next 100 units Rs 5 per Unit
After 200 Units Rs 10 per Unit
(For Example if input unit is 350 then total bill amount is 2000)

```python
try:
  x=float(input("enter the units consumed: "))
  try:
    if x<=100:
      print("Total bill amount is ",0)
    elif x<=200:
      print("Total bill amount is ",5*x)
    else:
      print("Ttoal bill amount is ",5*x+(x-200))
  except ValueError:
    print("Invalid input.")
except ValueError:
  print("Invalid Input")
```
